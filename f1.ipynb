```python
import json
import os
import requests
import pandas.io.json as pd_json
import pandas as pd
import re
from pandas.io.json import json_normalize
from pprint import pprint
```

```python
endpoint = 'https://ergast.com/api/f1/current.json'
res  = requests.get(endpoint)
res.ok
res
index_list = res.json()
# index_list.keys()
# pprint(index_list)
```

```python
races = res.json()['MRData']['RaceTable']['Races']
df = pd.DataFrame(races) 
df = json_normalize(races)
df.rename(columns = {'round':'Round','Circuit.circuitId':'circuitId','Circuit.url':'url','Circuit.circuitName':'Circuit Name',
                     'Circuit.Location.country':'Country', 'Circuit.Location.lat':'Latitude','Circuit.Location.long':'Longitude'}, inplace = True)
# df.head()
df[['Round','Circuit Name','Country','Latitude','Longitude']] 
```
